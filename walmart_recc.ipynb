```python
from bs4 import BeautifulSoup
def clean_and_split(text):
    if isinstance(text, str):
        soup = BeautifulSoup(text, "html.parser")
        cleaned_text = soup.get_text()
        return cleaned_text.split()
    else:

        return []
df['description']=df['description'].apply(clean_and_split)
```

```python
df.head()
```

```python
def to_list(row):
    if isinstance(row, str):
        cleaned_brand = row.replace(" ", "")
        return [cleaned_brand]
    else:
        return []
```

```python
df['brand']=df['brand'].apply(to_list)
```

```python
df['primary_category']=df['primary_category'].apply(lambda x: x.split())
df['sub_category_1']=df['sub_category_1'].apply(lambda x: x.split())
```

```python

df['sub_category_2']=df['sub_category_2'].apply(lambda x: x.split()if isinstance(x, str) else [])
```

```python
df
```

```python
df['tags']=df['brand']+df['description']+df['primary_category']+df['sub_category_1']+df['sub_category_2']
df
```

```python
df['tags']=df['tags'].apply(lambda x: ' '.join(x))
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000,stop_words='english',ngram_range=(1,3))
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
cv.vocabulary_
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
similarity = cosine_similarity(vector)
```

```python
type(similarity)
```

```python
df[df['title']=='Nuventin Retinol Cream'].index[0]
```

```python
def recommend(title):
    index = df[df['title']==title].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x: x[1])
    for i in distances[1:6]:
        print(df.iloc[i[0]].title)
```

```python
recommend('Nuventin Retinol Cream')
```
